## Overlays the model to experimental dataset 

```python
# Loading the required packages
import numpy as np   
import pandas as pd
import math                               
from scipy.integrate import odeint                        
import matplotlib.pyplot as plt                                 
%matplotlib inline
```

```python
def kinetics (y, t):
    
    #Tracked Species
    mRNA, template, T7RNAP = y
    
    #Equations

    dmRNAdt = k_trans*T7RNAP*template/(Km+template) - k_deg_mRNA * mRNA
    dtemplatedt = - k_deg_template*template
    dT7RNAPdt = -k_deg_T7RNAP*T7RNAP
   
    solutions = [dmRNAdt, dtemplatedt, dT7RNAPdt]
    
    return solutions
```

```python
y0_model2 = [0, 1, 0.21] #DoE_12 data
t_interval = np.linspace(0, 360, 5000)

#optimised parameters
k_trans = 7.03
Km = 0.937
k_deg_mRNA = 0.00433
k_deg_template = 0.00168
k_deg_T7RNAP = 0.000
```

```python
#Arrays for DoE_12 data
sol = odeint(func = kinetics, y0 = y0_model2, t = t_interval)
solved_2 = sol.transpose()
```

```python
time_siRNA = [0, 30, 60, 120, 180, 240, 300, 360]
DoE12_replicates = [
    [0.6820, 22.1558, 39.8414, 72.5187, 75.6872, 102.3556, 106.9185, 111.1474],
    [0.5906, 22.7643, 39.4630, 68.7246, 88.7178, 103.7814, 112.0117, 114.1678]
]

time_template = [0, 30, 60, 120, 180, 240, 300, 360]
template_replicates = [
    [0.9281, 0.719454107, 0.6861, 0.6445, 0.5091, 0.5209, 0.4875, 0.4959],
    [0.9057, 0.720755697, 0.6848, 0.6283, 0.5576, 0.5496, 0.5331, 0.5207]   
]
```

```python
# Calculate mean and standard deviation
DoE12_mean = np.mean(DoE12_replicates, axis=0)
DoE12_std = np.std(DoE12_replicates, axis=0)

template_DoE12_mean = np.mean(template_replicates, axis=0)
template_DoE12_std = np.std(template_replicates, axis=0)
```

```python
import matplotlib.pyplot as plt

# Plot mRNA only
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(t_interval, solved_2[0], color="red", label="Model")
ax.errorbar(time_siRNA, DoE12_mean, yerr=DoE12_std, fmt='o', color='black', label='Experimental dataset')
ax.set_title("IVT Kinetic Model for siRNA Synthesis")
ax.set_xlabel("Time (Min)")
ax.set_ylabel("siRNA concentration (µM)")
ax.legend()

plt.show()
```
